```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unknown columns

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
# Change weekly study hours column
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May","5-10")
```

```python
df.head()
```

```python
# Gender distribution
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From above chart we can conclude that number of females is more than number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
# Represent in Heat Map
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("The Relationship between parents education and students score")
plt.show()
```

```python
# From the above chart we have conclude that education of parents have good impact in students studies
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot = True)
plt.show()
```

```python
# From the above chart we have conclude that there is no/neglicible impact on students score due to parents martial status
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.show()
```
